```python
# To use built in point and Polygon
# you may have to install the library using "pip install shapely"
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon


# Function telling if points lie inside polygon or not
# returns a list of True/False against each point
def if_lies_inside_polygon(polygon, points):
    # Extract coordinates from input dict
    polygon_coordinates = polygon[0]["coordinates"].split(",")
    polygon = []
    for i in range(len(polygon_coordinates)):
        if i%2 == 0:
            polygon.append((float(polygon_coordinates[i]), float(polygon_coordinates[i+1])))
    polygon = Polygon(polygon)
    print(polygon)
    result = []
    for point in points:
        point = Point(point["long"], point["lat"])
        if polygon.contains(point):
            result.append(True)
        else:
            result.append(False)
    return result

# Sample input polygon and points dicts
polygon = [
    {
        "geoSearchType":"draw",
         "coordinates":"50.131279526346844,8.254234375000005,49.421832443657046,8.045494140625005,49.12795317133869,8.583824218750005,49.53604229195023,9.243003906250005,49.87707489536068,8.649742187500005"
    }
]
points = [
    {id:1, "long": 49.32083,"lat": 8.43111},
    {id:2, "long": 50.131279526346844,"lat": 9.254234375000005},
    {id:3, "long": 50.031279526346844,"lat": 7.254234375000005},
]

# Function call
result = if_lies_inside_polygon(polygon, points)
print(result)
    
# Code Credits
# ayeshasiddiqatheprogrammer@gmail.com
```
